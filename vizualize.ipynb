# Partie 2 - Exploration et visualisation des données

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
df = pd.read_csv('weather_data_1000_cleaned.csv')
```

# Etape 1 - Histogramme des températures

```python
plt.hist(df['Temperature'], bins=20, color='blue')
plt.title('Histogram of Temperature')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
```

# Etape 2 - Boxplot des précipitations

```python
sns.boxplot(df, y=df['Precipitation'])
```

# Etape 3 - Relation entre température et précipitations

```python
df['date'] = pd.to_datetime(df['Date'])
plt.figure(figsize=(8, 4))
sns.scatterplot(data=df, x='Precipitation', y='Temperature', hue='Temperature', palette='coolwarm')
sns.regplot(scatter=False, color='red', data=df, y='Temperature', x='Precipitation')
```

# Etape 4 - Analyse temporelle des températures

```python
df['Date'] = pd.to_datetime(df['Date'])
plt.title('Evolution of Temperatures and Precipitation')
plt.scatter(data=df, x='Date', y='Temperature',color='r')
plt.scatter(data=df, x='Date', y='Precipitation',color='b')
```

# Etape 5 - Heatmap des corrélations

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True)
```

# Etape 6 - Graphiques interactifs avec Plotly

```python
import plotly.express as px
px.scatter(data_frame=df, x='Date', y='Temperature', color='Temperature')
```
